```python
import pandas as pd
df=pd.read_csv('/content/framingham.csv')
df
```

```python
df.isnull().sum()
```

```python
x=df['cigsPerDay'].mode()[0]
x
df['cigsPerDay']=df['cigsPerDay'].fillna(df['cigsPerDay'].mode()[0])
df['BPMeds']=df["BPMeds"].fillna(df['BPMeds'].mode()[0])
df['totChol']=df['totChol'].fillna(df['totChol'].mean())
df['BMI']=df['BMI'].fillna(df['BMI'].mean())
df['heartRate']=df['heartRate'].fillna(df['heartRate'].mean())
df['glucose']=df['glucose'].fillna(df['glucose'].mean())
df.drop(['education'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
X.columns
```

```python
#feature selection
#selection of specific features that is reqiured for output
#feature engineering-selecting a new feature from existing features 
#+ve correlation-when one feature increases,another feature increases
    #demand  price
#-ve correlation-when one feature increases, another feature decreases
     #supply  demand
  
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
test=SelectKBest(score_func=chi2)
fi=test.fit(X,y)
fi.scores_
```

```python
col=X.columns
score=pd.DataFrame({'features':col,'score_chi2':fi.scores_})
score
```

```python
score.sort_values(by='score_chi2',ascending=False)
```

```python
#we can select highly relevant 10 features
X=df[['sysBP','glucose','age','totChol','cigsPerDay','diaBP','prevalentHyp','diabetes','BPMeds']]
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
