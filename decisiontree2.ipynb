```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
dic=load_iris()
dic
```

```python
X=dic.data
X
```

```python
y=dic.target
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier(criterion='entropy')
dec.fit(X_train,y_train)
```

```python
y_pred=dec.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))#correct predictions are diagonally placed
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
tree.plot_tree(dec,feature_names= ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'],filled=True)
```
