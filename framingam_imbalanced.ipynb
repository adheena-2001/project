```python
import pandas as pd
df=pd.read_csv('/content/framingham.csv')
df
```

```python

df['TenYearCHD'].value_counts()
```

```python
#we can check if a data is imbalanced using
#1.oversampling-increase number of minority samples equals to number of majority samples
#2.undersampling-decrease number of majority samples to number of minority samples
```

```python
df.isnull().sum()
```

```python
df.drop(['education'],axis=1,inplace=True)
df
```

```python
x=df['cigsPerDay'].mode()[0]
x
```

```python
df['cigsPerDay']=df['cigsPerDay'].fillna(df['cigsPerDay'].mode()[0])
```

```python
df['BPMeds']=df["BPMeds"].fillna(df['BPMeds'].mode()[0])
```

```python
df['totChol']=df['totChol'].fillna(df['totChol'].mean())
```

```python
df['BMI']=df['BMI'].fillna(df['BMI'].mean())
```

```python
df['heartRate']=df['heartRate'].fillna(df['heartRate'].mean())
```

```python
df['glucose']=df['glucose'].fillna(df['glucose'].mean())
```

```python
df.isnull().sum()
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
#1.oversampling
from imblearn.over_sampling import SMOTE#SMOTE-synthetic minority oversampling technique
oversample=SMOTE()
Xo,yo=oversample.fit_resample(X,y)
```

```python
yo.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Xo,yo,test_size=0.3,random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train_new=sc.transform(X_train)
X_train_new
```

```python
X_test_new=sc.transform(X_test)
X_test_new
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_new,y_train)
y_pred=knn.predict(X_test_new)
y_pred
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
#undersampling
from imblearn.under_sampling import RandomUnderSampler
under=RandomUnderSampler()
Xu,yu=under.fit_resample(X,y)
```

```python
yu.value_counts()
```

```python
Xu
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Xu,yu,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train_new=sc.transform(X_train)
X_train_new
```

```python
X_test_new=sc.transform(X_test)
X_test_new
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_new,y_train)
y_pred=knn.predict(X_test_new)
y_pred
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
