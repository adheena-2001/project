```python
import pandas as pd
df=pd.read_csv('/content/Churn_Modelling.csv')
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['Exited'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Surname']=le.fit_transform(df['Surname'])
df['Geography']=le.fit_transform(df['Geography'])
df['Gender']=le.fit_transform(df['Gender'])
df
```

```python
df.dtypes
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
Xo,yo=oversample.fit_resample(X,y)
```

```python
yo.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Xo,yo,test_size=0.3,random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train_new=sc.transform(X_train)
X_train_new
```

```python
X_test_new=sc.transform(X_test)
X_test_new
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_new,y_train)
y_pred_knn=knn.predict(X_test_new)
y_pred_knn
```

```python
y_test
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred_knn))
print(classification_report(y_test,y_pred_knn))
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train_new,y_train)
y_pred_nb=nb.predict(X_test_new)
y_pred_nb
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred_nb))
print(classification_report(y_test,y_pred_nb))
```

```python
from sklearn.svm import SVC
sv=SVC(kernel='linear')
sv.fit(X_train_new,y_train)
y_pred_sv=sv.predict(X_test_new)
print(y_pred_sv)
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred_sv))
print(classification_report(y_test,y_pred_nb))
```

```python
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier(criterion='entropy')
dec.fit(X_train_new,y_train)
y_pred_dec=dec.predict(X_test_new)
y_pred_dec
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_pred_dec,y_test))
print(classification_report(y_pred_dec,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100,criterion='gini')
clf.fit(X_train_new,y_train)
```

```python
y_pred_rf=clf.predict(X_test_new)
y_pred_rf
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_pred_rf,y_test))
print(classification_report(y_pred_rf,y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_clf=GradientBoostingClassifier(n_estimators=100,learning_rate=1.0,max_depth=1,random_state=0)
gb_clf.fit(X_train_new,y_train)
gb_pred=gb_clf.predict(X_test_new)
gb_pred
```

```python
print(ConfusionMatrixDisplay.from_predictions(gb_pred,y_test))
print(classification_report(gb_pred,y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf=AdaBoostClassifier(n_estimators=100)
clf.fit(X_train_new,y_train)
y_pred_ab=clf.predict(X_test_new)
y_pred_ab
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_pred_ab,y_test))
print(classification_report(y_pred_ab,y_test))
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(X_train_new,y_train)
y_pred_lg=reg.predict(X_test_new)
y_pred_lg
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_pred_ab,y_test))
print(classification_report(y_pred_ab,y_test))
```
